```python
import pandas as pd
```

```python
c = twint.Config()

c.Geo = '40.6652,-73.9125, 1.6km'
#c.Search = '(covid OR covid-19 OR covid19 OR Coronavirus OR Corona OR covid-20 OR covid20 OR Virus OR coronavir)'
#c.Search = '(virus OR covid19 OR covid OR corona OR coronaviras OR corona-virus OR covid19-virus OR sarscov2 OR #covid19 OR #covid OR #corona OR #coronaviras OR #corona-virus OR #covid19-virus OR #sarscov2)'
c.Since = '2020-03-01 00:00:00'
c.Until = '2020-07-01 00:00:00'
c.Output = 'Brownsville_march_june.csv'
c.Store_csv = True

# run and store in a dataframe
twint.run.Search(c)
```

```python
Brownsville = pd.read_csv('Brownsville_march_june.csv')
print(Brownsville.shape)
```
